```python
import pandas as pd

# Replace this with your actual path
folder_path = "/Users/santiago/Desktop/CS/ML_Lab7Folder"

# List of exact filenames
file_names = [
    "business_data.csv",
    "education_data.csv",
    "entertainment_data.csv",
    "sports_data.csv",
    "technology_data.csv"
]

# Read and combine
dataframes = [pd.read_csv(f"{folder_path}/{file}") for file in file_names]
data = pd.concat(dataframes, ignore_index=True)

# Quick checks
print(f"Total samples: {len(data)}")
print(data['category'].value_counts())
```

The dataset used in this project consists of news articles from The Indian Express, grouped into five topical categories: Business, Education, Entertainment, Sports, and Technology. Each category was originally stored in a separate CSV file, each containing 2,000 articles. These files were business, education, entertainment, sports, and technology.csv. All five files were loaded and merged into a single DataFrame, resulting in a combined dataset of 10,000 text samples, evenly distributed across the five classes. In each we find 5 rows Headlines, Description, Content, URL, and category. The Content column is used as the input feature, and the category column is used as the class label.

### Preprocessing 

```python
from sklearn.preprocessing import LabelEncoder

# Remove rows with missing content
data.dropna(subset=["content", "category"], inplace=True)

# Reset index
data.reset_index(drop=True, inplace=True)

# Encode labels into integers
label_encoder = LabelEncoder()
data['label'] = label_encoder.fit_transform(data['category'])

# View encoded classes
print(label_encoder.classes_)
```

To clean up the data before training, I started by dropping any rows that had missing values in either the content or category columns as we only want to work with complete examples. After that, we reset the index so everything stays in order. Then I encoded the category labels into numbers using LabelEncoder, since models can’t work directly with string labels. This turned each category into an integer from 0 to 4, where Business became 0, Education was 1, and so on. This lets me treat the task as a multi-class classification problem with five total classes.

### Tokenize

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

# Use full content text
texts = data['content'].astype(str).tolist()
labels = data['label'].values

# Tokenize text
tokenizer = Tokenizer(num_words=20000, oov_token="<OOV>")
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)

# Set max sequence length (you can tweak this later)
MAX_LEN = 300
X = pad_sequences(sequences, maxlen=MAX_LEN, padding='post', truncating='post')
y = labels
```

After encoding the labels, I preprocessed the article content so it could be used as input for the models. I started by converting all the content into plain strings and made a list out of them. Then I used TensorFlow’s Tokenizer to turn the text into sequences of numbers. I limited the tokenizer to the top 20,000 most common words and used an <OOV> token for anything outside that range. After fitting the tokenizer on the dataset, I converted each article into a sequence of word indices. Since deep learning models need inputs to be the same length, I set a max sequence length of 300 and padded or truncated each sequence to match that. The result was a clean X matrix of shape (10000, 300) that I can feed into the model, and a y array holding the numeric category labels.



## 1.2 Evaluation Metrics 

For evaluating the model, I’m using macro F1-score instead of just going with accuracy. Since the dataset is perfectly balanced across all five categories, macro F1 makes more sense because it gives equal weight to each class by calculating F1 for each one individually and then averaging them. Accuracy can be misleading because it only tells you how often the model is right overall, not whether it’s performing equally well across every category.

In this case, I want to make sure the model doesn’t just do well on one or two classes while doing poorly on others, especially since something like misclassifying a tech article as business could actually matter a lot in practice. F1-score takes both precision and recall into account, so I get a better idea of how well the model is handling each class, not just how many it gets right. Precision is also helpful on its own to see how often the model's predictions are actually correct for each category, which gives more insight into where it's being too confident or too conservative.



## 1.3 Data Split 

Originally, we considered using Stratified 10-Fold Cross Validation since it gives a more complete picture of model performance across different splits. But after testing it out, we realized it was too slow and computationally expensive for our setup, especially given the size of the data and the models we’re running.

Instead, we decided to use a more efficient 80/20 stratified train/test split. This still preserves the balance across all five classes in both sets, but it’s much faster to work with. For our project, where we mainly want to compare model performance and get reliable results without overcomplicating the training process, this approach made the most sense.

```python
import matplotlib.pyplot as plt

def plot_training_history(history, title="Model Accuracy and Loss"):
    plt.figure(figsize=(12, 4))
    
    # Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train')
    plt.plot(history.history['val_accuracy'], label='Validation')
    plt.title('Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    # Loss
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train')
    plt.plot(history.history['val_loss'], label='Validation')
    plt.title('Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.suptitle(title)
    plt.show()

plot_training_history(history_cnn, title="CNN Baseline")
```

For our first CNN model, we used the Functional API and kept the architecture pretty straightforward. We started with an embedding layer that maps each word to a 100-dimensional vector. After that, we used a Global Max Pooling layer to compress everything down to the most important signals. We followed that with a 0.5 dropout to help reduce overfitting, and added a dense layer with 64 units before finishing with a softmax output layer for 5-class classification.

The model had around 2 million trainable parameters. During training, it performed really well as the validation accuracy hit around 97.8% by the end of 10 epochs and the loss dropped significantly. The plots show that both the accuracy and loss curves stabilized nicely, which means the model converged and didn’t overfit too much.

### Transformer-based

#### Transformer Encoder Block Definition

```python
from tensorflow.keras.layers import LayerNormalization, MultiHeadAttention, Dense, Dropout, Add
from tensorflow.keras import layers

def transformer_encoder(inputs, head_size, num_heads, ff_dim, dropout=0):
    # Multi-head self attention
    x = MultiHeadAttention(key_dim=head_size, num_heads=num_heads)(inputs, inputs)
    x = Dropout(dropout)(x)
    x = Add()([x, inputs])
    x = LayerNormalization(epsilon=1e-6)(x)

    # Feed-forward layer
    x_ff = Dense(ff_dim, activation='relu')(x)
    x_ff = Dense(inputs.shape[-1])(x_ff)
    x_ff = Dropout(dropout)(x_ff)
    x = Add()([x_ff, x])
    x = LayerNormalization(epsilon=1e-6)(x)
    return x
```

#### Transformer Model Definition

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, GlobalAveragePooling1D

# Input
inputs = Input(shape=(MAX_LEN,))

# Embedding
embedding_layer = Embedding(input_dim=VOCAB_SIZE, output_dim=EMBEDDING_DIM)(inputs)

# Transformer Encoder Block
x = transformer_encoder(embedding_layer, head_size=64, num_heads=4, ff_dim=128, dropout=0.1)

# Global average pooling
x = GlobalAveragePooling1D()(x)
x = Dropout(0.3)(x)
x = Dense(64, activation='relu')(x)
outputs = Dense(NUM_CLASSES, activation='softmax')(x)

transformer_model = Model(inputs, outputs)

transformer_model.compile(loss='sparse_categorical_crossentropy',
                          optimizer=Adam(learning_rate=1e-3),
                          metrics=['accuracy'])

transformer_model.summary()
```

#### Training the Transformer Base

```python
history_transformer = transformer_model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=10,
    batch_size=64
)
```

#### Plotting

```python
plot_training_history(history_transformer, title="Transformer Baseline")
```

For our first Transformer model, we used a basic setup with one multi-head attention layer (4 heads), followed by layer normalization and a feed-forward block. We kept the embedding size at 100 to stay consistent with the CNNs, and used global average pooling before the dense layers. Overall, the model had just over 2.1 million parameters.

Training went really well with the model hitting around 98% validation accuracy by epoch 2 and stayed strong throughout. Training accuracy hit 100% and the loss dropped super low. ### Transformer Variation 

####  Transformer Encoder Block

```python
# Use the same transformer_encoder block
def transformer_encoder(inputs, head_size, num_heads, ff_dim, dropout=0):
    x = MultiHeadAttention(key_dim=head_size, num_heads=num_heads)(inputs, inputs)
    x = Dropout(dropout)(x)
    x = Add()([x, inputs])
    x = LayerNormalization(epsilon=1e-6)(x)

    x_ff = Dense(ff_dim, activation='relu')(x)
    x_ff = Dense(inputs.shape[-1])(x_ff)
    x_ff = Dropout(dropout)(x_ff)
    x = Add()([x_ff, x])
    x = LayerNormalization(epsilon=1e-6)(x)
    return x
```

#### Transformer Variant #2

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, GlobalAveragePooling1D, Dropout, Dense

# Input layer
inputs = Input(shape=(MAX_LEN,), name="input_layer")

# Embedding
x = Embedding(input_dim=VOCAB_SIZE, output_dim=EMBEDDING_DIM, name="embedding")(inputs)

# Transformer block with 8 heads
x = transformer_encoder(x, head_size=64, num_heads=8, ff_dim=128, dropout=0.1)

# Output
x = GlobalAveragePooling1D()(x)
x = Dropout(0.3)(x)
x = Dense(64, activation='relu')(x)
outputs = Dense(NUM_CLASSES, activation='softmax')(x)

# Model
transformer_model_8h = Model(inputs, outputs, name="transformer_8_heads")

transformer_model_8h.compile(loss='sparse_categorical_crossentropy',
                             optimizer=Adam(learning_rate=1e-3),
                             metrics=['accuracy'])

transformer_model_8h.summary()
```

#### Training Transformer Model #2

```python
history_transformer_8h = transformer_model_8h.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=10,
    batch_size=64
)
```

#### Plotting

```python
plot_training_history(history_transformer_8h, title="Transformer Variant (8 Attention Heads)")
```

For this second Transformer, we kept the same setup as before but bumped the number of attention heads from 4 to 8. The goal was to see if giving the model more heads to focus on different parts of the input would help it catch more patterns. That change brought the parameter count up to about 2.2 million.

Performance-wise, it held up super well. Validation accuracy hit over 98% just like the first model, and it stayed consistent across all 10 epochs. Training accuracy reached 100% again, and loss dropped close to zero. Compared to the first Transformer, the validation loss here stayed a bit lower overall, which makes me think it might generalize just a little better.

The accuracy curve basically plateaued by the second or third epoch, and both training and validation loss stayed low and steady. From the graphs, it looks like this model trained really clean and didn’t overfit.
